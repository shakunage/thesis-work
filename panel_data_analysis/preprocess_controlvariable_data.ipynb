```python
from datetime import datetime
import openpyxl
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
```

## Preprocess OMXN40 Control Variable Data

Convert the OMXN40 index data from Excel to CSV format matching the Yahoo Finance output structure.

```python
# Load the OMXN40 data from Excel
input_file = Path('../scraping/output_data/control_variables/EODHist_20120101-20251231_OMXN40.xlsx')
output_file = Path('../scraping/output_data/yahoofinance_^OMXN40.csv')

print(f"Loading data from: {input_file}")

# Read the Excel file
df_omxn40 = pd.read_excel(input_file)

print(f"Original shape: {df_omxn40.shape}")
print(f"Original columns: {df_omxn40.columns.tolist()}")
print(f"\nFirst few rows:")
df_omxn40.head()
```

```python
# Rename Index Value to Close
df_omxn40 = df_omxn40.rename(columns={'Index Value': 'Close'})
df_omxn40 = df_omxn40.rename(columns={'Trade Date': 'Date'})

# Select Date, High, Low, and Close columns (removing Net Change)
df_processed = df_omxn40[['Date', 'High', 'Low', 'Close']].copy()

# Round numeric values to 2 decimal places
df_processed['High'] = df_processed['High'].round(2)
df_processed['Low'] = df_processed['Low'].round(2)
df_processed['Close'] = df_processed['Close'].round(2)

print(f"Processed shape: {df_processed.shape}")
print(f"Processed columns: {df_processed.columns.tolist()}")
print(f"\nSample of processed data:")
print(df_processed.head(10))
print(f"\nData types:")
print(df_processed.dtypes)
print(f"\nDate range:")
print(f"  From: {df_processed['Date'].min()}")
print(f"  To:   {df_processed['Date'].max()}")
```

```python
# Save to CSV
df_processed.to_csv(output_file, index=False)

print(f"✓ Successfully saved to: {output_file}")
print(f"  Total rows: {len(df_processed):,}")
print(f"  File size: {output_file.stat().st_size / 1024:.1f} KB")
```
