```python
import pandas as pd
```

```python
# Import FinnSentiment-1.1.tsv using polars
df = pd.read_csv('../data/finsen-v1-1-src/FinnSentiment-1.1.tsv', sep='\t', header=None)
df.head()
```

```python
df.info()
```

```python
df = df[df[0] == df[1]]
```

```python
df = df[df[1] == df[2]]
```

```python
df.info()
```

```python
# Keep only columns at indices 3, 19, and 20
df_trimmed = df.iloc[:, [2, 19]]
df_trimmed.head(25)
```

```python
# Rename columns "2" and "19" to "label" and "text"
df_trimmed = df_trimmed.rename(columns={2: "label", 19: "text"})
```

```python
label_map = {-1: 0, 0: 1, 1: 2}
df_trimmed["label"] = df_trimmed["label"].map(label_map)
```

```python
# Save df_trimmed as a pickle file
df_trimmed.to_pickle('../data/finnsentiment_cleaned_unambiguous.pkl')
```
